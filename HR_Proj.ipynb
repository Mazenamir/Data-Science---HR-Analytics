```python

fig,ax = plt.subplots(2,2, figsize=(10,10))
plt.suptitle("The distrubtion",fontsize = 20)


figure_1 = sns.distplot(df['Age'], ax = ax[0,0] )
ax[0][0].set_title('Distribution of AGE')

figure_2 = sns.distplot(df['DailyRate'] ,ax = ax[0,1])
ax[0][1].set_title('Distribution of DailyRate')

figure_3 =sns.distplot(df['YearsAtCompany'], ax = ax[1,0])
ax[1][0].set_title('Distribution of  Years at company')

figure_4 = sns.distplot(df['MonthlyIncome']  , ax = ax[1,1])
ax[1][1].set_title('Distribution of MonthlyIncome')



plt.show()
```

```python
fig1, ax1 = plt.subplots(2, 1, figsize=(10, 10))
attrition_counts = df['Attrition'].value_counts()

bar_1 = sns.barplot(y=attrition_counts.index, x=attrition_counts.values, ax=ax1[0])

# Adding values to the bars in the first bar plot
for index, value in enumerate(attrition_counts.values):
    bar_1.text(value, index, str(value), ha='left', va='center')

ax1[0].set_title('Distribution of Attrition')

gender_counts = df['Gender'].value_counts()
bar_2 = sns.barplot(y=gender_counts.index, x=gender_counts.values, ax=ax1[1])

# Adding values to the bars in the second bar plot
for index, value in enumerate(gender_counts.values):
    bar_2.text(value, index, str(value), ha='left', va='center')

ax1[1].set_title('Distribution of Gender')

plt.show()
```
